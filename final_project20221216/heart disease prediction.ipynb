```python
# 查看数据范围 判断离散or连续
categorical_val = []
continous_val = []
for column in df.columns:
    print('==============================')
    print(f"{column} : {df[column].unique()}") # 列数据的唯一值
    if len(df[column].unique()) <= 10:
        categorical_val.append(column)
    else:
        continous_val.append(column)
```

```python
# 制作数据表格
datainf = pd.DataFrame(columns=['Name','Description','Scale'])
Des = ['年龄','性别','胸痛','静息血压','胆固醇','空腹血糖','静息心电图结果','最大心率',
       '运动性心绞痛','相对于体重而言，运动所引起的压抑感','峰值运动段的斜率','荧光透视染色的主要血管数量','铊压力结果','是否患心脏病']
categorical_ = []
continous_ = []
for i, c in enumerate(df.columns, 0):
    if c in categorical_val:
        scale = " ".join(map(str,df[c].unique()))
        categorical_.append(Des[i])
    else:
        scale = " ".join([str(min(df[c].unique())),'~',str(max(df[c].unique()))])
        continous_.append(Des[i])
    item = {'Name':c,'Description':Des[i],'Scale':scale}
    datainf.loc[i] = item
datainf
```

```python
print('、'.join(categorical_))
print('、'.join(continous_))
```

```python
# Data Processing
# 离散属性one-hot编码 连续属性标准化

categorical_val.remove('target')
dataset = pd.get_dummies(df, columns = categorical_val)

from sklearn.preprocessing import StandardScaler

s_sc = StandardScaler()
col_to_scale = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']
dataset[col_to_scale] = s_sc.fit_transform(dataset[col_to_scale])
```

```python
databack = dataset
```

```python
# 打印模型评估报告 混淆矩阵
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

def print_score(clf, X_train, y_train, X_test, y_test, train=True):
    if train:
        pred = clf.predict(X_train)
        clf_report = pd.DataFrame(classification_report(y_train, pred, output_dict=True)) # 模型评估报告
        print("Train Result:\n================================================")
        print(f"Accuracy Score: {accuracy_score(y_train, pred) * 100:.2f}%")
        print("_______________________________________________")
        print(f"CLASSIFICATION REPORT:\n{clf_report}")
        print("_______________________________________________")
        print(f"Confusion Matrix: \n {confusion_matrix(y_train, pred)}\n")

    elif train==False:
        pred = clf.predict(X_test)
        clf_report = pd.DataFrame(classification_report(y_test, pred, output_dict=True))
        print("Test Result:\n================================================")
        print(f"Accuracy Score: {accuracy_score(y_test, pred) * 100:.2f}%")
        print("_______________________________________________")
        print(f"CLASSIFICATION REPORT:\n{clf_report}")
        print("_______________________________________________")
        print(f"Confusion Matrix: \n {confusion_matrix(y_test, pred)}\n")
```

```python
# 分割数据集 7:3
from sklearn.model_selection import train_test_split

X = dataset.drop('target', axis=1)
y = dataset.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
print(len(X_train))
print(len(X_test))
```

```python
# logistic regression model
from sklearn.linear_model import LogisticRegression

lr_clf = LogisticRegression(solver='liblinear')
lr_clf.fit(X_train, y_train)

print_score(lr_clf, X_train, y_train, X_test, y_test, train=True)
print_score(lr_clf, X_train, y_train, X_test, y_test, train=False)
```

```python
test_score = accuracy_score(y_test, lr_clf.predict(X_test)) * 100
train_score = accuracy_score(y_train, lr_clf.predict(X_train)) * 100

results_df = pd.DataFrame(data=[["Logistic Regression", train_score, test_score]],
                          columns=['Model', 'Training Accuracy %', 'Testing Accuracy %'])
results_df
```

```python
from sklearn.decomposition import PCA
score_test = []
score_train = []
for i in range(4,22):
    pca_breast = PCA(n_components=i)
    data = pca_breast.fit_transform(dataset)

    X = data
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
    lr_clf = LogisticRegression(solver='liblinear')
    lr_clf.fit(X_train, y_train)
    test_score = accuracy_score(y_test, lr_clf.predict(X_test)) * 100
    train_score = accuracy_score(y_train, lr_clf.predict(X_train)) * 100
    score_test.append(test_score)
    score_train.append(train_score)
print(f"Accuracy Test", list(map(lambda x:round(x,2) , score_test)) )
print(f"Accuracy Train", list(map(lambda x:round(x,2) , score_train)) )
plt.figure(figsize=(10, 8))
plt.plot(range(4,22),score_test,c="salmon")
plt.plot(range(4,22),score_train,c="lightblue")
plt.legend(["Test", "Train"])
plt.show()
```
